# 游댍 DataWranling enfocado a MachineLearning

En esta ocasion, se han realizado los ajustes de limpieza necesarios para dejar los datos listos para la aplicaci칩n de algoritmos de aprendizaje autom치tico. Detallaremos los cambios a continuaci칩n: <br>
Detallaremos los cambios a continuaci칩n:<br>
* Se ha aplicado un modelo de codificaci칩n One Hot.
* Sobre la residencia del empleado, hemos agrupado las categor칤as en [US] y [NO US], representando las observaciones que est치n dentro y fuera de Estados Unidos por motivos de equilibrio.
* Sobre el tipo de contrato, hemos agrupado las categor칤as [Full_time] y [No_FT], representando las observaciones que est치n a tiempo completo y las que no.
* Al ver la cantidad de observaciones en la categor칤a de fulltime, he decidido no utilizar estas categor칤as, dado que no representan nada en mi conjunto de datos; el 94% de las observaciones est치n bajo Full_Time.
* Se ha generado la columna menos_experiencia y mas_experiencia para agrupar las categor칤as [Senior - Ejecutivo] y [Junior - Mid_Junior] respectivamente, para ayudar con el equilibrio de los datos.

### One hot encoder y arreglos

````python
df_salaryML_OneHotEncoder = df_salary.copy()
````

#### Residencia Empleado.

````python
df_salaryML_OneHotEncoder['Residencia_del_empleado'] = df_salaryML_OneHotEncoder['Residencia_del_empleado'].apply(lambda x: x if x == 'US' else 'No_US')
````

#### Tipo de contratacion


````python
df_salaryML_OneHotEncoder['Tipo_de_contrato'] = df_salaryML_OneHotEncoder['Tipo_de_contrato'].apply(lambda x: x if x == 'Full_Time' else 'No_FT')
````

#### mas y menos experiencia

````python
df_salaryML_OneHotEncoder['mas_experiencia'] = np.where((df_salaryML_OneHotEncoder['Nivel_de_experiencia'] == 'Senior') | (df_salaryML_OneHotEncoder['Nivel_de_experiencia'] == 'Ejecutivo'), 1, 0)
df_salaryML_OneHotEncoder['menos_experiencia'] = np.where((df_salaryML_OneHotEncoder['Nivel_de_experiencia'] == 'Mid_Junior') | (df_salaryML_OneHotEncoder['Nivel_de_experiencia'] == 'Junior'), 1, 0)
df_salaryML_OneHotEncoder = df_salaryML_OneHotEncoder.drop('Nivel_de_experiencia', axis=1)
````

#### Anio laboral

````python
df_salaryML_OneHotEncoder['Anio_2022'] = np.where(df_salaryML_OneHotEncoder['Anio_Laboral'] == 2022, 1, 0)
df_salaryML_OneHotEncoder['Anterior_2022'] = np.where(df_salaryML_OneHotEncoder['Anio_Laboral'] < 2022, 1, 0)
````

````python
df_salaryML_OneHotEncoder.tail()
````

#### OneHotEncoder

````python
from sklearn.preprocessing import OneHotEncoder

# Crear una instancia de OneHotEncoder
encoderOHE = OneHotEncoder()

# Ajustar y transformar los datos
encoded_data = encoderOHE.fit_transform(df_salaryML_OneHotEncoder[['Tipo_de_contrato', 'Residencia_del_empleado', 'Tamanio_de_compania']])

# Obtener las categor칤as 칰nicas para cada columna
feature_names = encoderOHE.get_feature_names_out(['Tipo_de_contrato', 'Residencia_del_empleado', 'Tamanio_de_compania'])

# Convertir el resultado a un DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=feature_names)

# Concatenar el DataFrame codificado con el DataFrame original
df_encoded = pd.concat([df_salaryML_OneHotEncoder, encoded_df], axis=1)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import r2_score, mean_absolute_error

# Selecionar las columnas para el PCA
columns_for_pca = ['mas_experiencia', 'menos_experiencia', 'Anio_2022', 'Anterior_2022','Residencia_del_empleado_No_US', 'Residencia_del_empleado_US']  # Reemplaza con las columnas que deseas utilizar en el PCA

# Separar la variable objetivo (salario) y las caracter칤sticas
X = df_encoded[columns_for_pca]
y = df_encoded['Salario_en_dolares']

# Normalizar los datos antes de aplicar PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Obtener la varianza explicada acumulada
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_cumulative = explained_variance_ratio.cumsum()

# Graficar la varianza explicada acumulada
plt.plot(range(1, len(explained_variance_cumulative) + 1), explained_variance_cumulative, marker='o', linestyle='--')
plt.xlabel('N칰mero de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Varianza explicada acumulada por componente principal')
plt.grid(True)
plt.show()

# Seleccionar el n칰mero de componentes principales a utilizar
n_components = 3  # Reemplaza con el n칰mero de componentes principales que deseas utilizar

# Reducir la dimensionalidad a trav칠s del PCA
X_selected = X_pca[:, :n_components]

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Crear el modelo de XGBoost
model = xgb.XGBRegressor(colsample_bytree= 0.8, learning_rate= 0.01, max_depth= 3, n_estimators= 500, subsample= 1.0)

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de entrenamiento y prueba
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calcular las m칠tricas de evaluaci칩n en los datos de entrenamiento
r2_train = r2_score(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)

# Calcular las m칠tricas de evaluaci칩n en los datos de prueba
r2_test = r2_score(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

# Imprimir las m칠tricas
print('M칠tricas en los datos de entrenamiento:')
print('R:', r2_train)
print('MAE:', mae_train)
print('\nM칠tricas en los datos de prueba:')
print('R:', r2_test)
print('MAE:', mae_test)
````
